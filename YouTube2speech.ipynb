```python
import yt_dlp
import moviepy.editor as mp
import requests
import os

# Deepgram API Key and Eleven Labs API Key (Replace with your keys)
DEEPGRAM_API_KEY = 'PLEASE USE YOU DEEPGRAM API KEY HERE'
ELEVEN_LABS_API_KEY = 'PLEASE USE ELEVEN LABS API KEY HERE'

# Function to download YouTube video
def download_youtube_video(url, download_path):
    """Downloads a YouTube video to the specified path using yt_dlp."""
    ydl_opts = {
    'format': 'bestvideo+bestaudio/best',
    'outtmpl': os.path.join(download_path, 'video.%(ext)s'),  # Save the video with the correct extension
    'merge_output_format': 'mp4'  # Ensure the final file is in mp4 format
}

    try:
        with yt_dlp.YoutubeDL(ydl_opts) as ydl:
            ydl.download([url])
        print("Download successful!")
    except Exception as e:
        print(f"Error downloading video: {e}")

# Function to extract audio from video
def extract_audio_from_video(video_path, audio_output_path):
    """Extracts the audio from a video file and saves it as a WAV file."""
    try:
        # Ensure you are referencing the correct video file
        video_clip = mp.VideoFileClip(video_path)
        audio_clip = video_clip.audio
        audio_clip.write_audiofile(audio_output_path)
        print(f"Audio extracted and saved to: {audio_output_path}")
    except OSError as e:
        print(f"Error: {e}")
        print(f"Please check if the video file exists at: {video_path}")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

# Function to transcribe audio using Deepgram API
def transcribe_audio_deepgram(audio_path):
    """Sends the audio file to Deepgram API for transcription and returns the transcribed text."""
    url = "https://api.deepgram.com/v1/listen"
    headers = {
        "Authorization": f"Token {DEEPGRAM_API_KEY}",
        "Content-Type": "audio/wav"
    }
    with open(audio_path, 'rb') as audio_file:
        response = requests.post(url, headers=headers, data=audio_file)
    if response.status_code == 200:
        transcription = response.json().get('results', {}).get('channels', [])[0].get('alternatives', [])[0].get('transcript', '')
        print("Transcription successful!")
        return transcription
    else:
        print(f"Error transcribing audio: {response.json()}")
        return None

# Function to generate speech using Eleven Labs API
def generate_speech_eleven_labs(text, output_audio_path):
    """Uses Eleven Labs API to generate speech from text."""
    # Replace 'your_voice_id' with the actual voice ID you have access to
    url = "https://api.elevenlabs.io/v1/text-to-speech/9BWtsMINqrJLrRacOk9x"
    headers = {
        "xi-api-key": ELEVEN_LABS_API_KEY,
        "Content-Type": "application/json"
    }
    data = {
        "text": text
    }
    response = requests.post(url, headers=headers, json=data)
    if response.status_code == 200:
        with open(output_audio_path, "wb") as audio_file:
            audio_file.write(response.content)
        print(f"Audio generated and saved to {output_audio_path}")
    else:
        print(f"Error generating speech: {response.json()}")




# Main function
if __name__ == "__main__":
    # Step 1: Download YouTube video
    video_url = input("Enter YouTube video URL: ")
    download_path = "./"  # Directory where video will be saved
    download_youtube_video(video_url, download_path)

    # Step 2: Extract audio from the downloaded video
    video_file_path = os.path.join(download_path, "video.mp4")  # The expected output file after merging
    audio_file_path = os.path.join(download_path, "audio.wav")
    extract_audio_from_video(video_file_path, audio_file_path)

    # Step 3: Convert audio to text using Deepgram API
    transcription = transcribe_audio_deepgram(audio_file_path)

    if transcription:
        print(f"Transcription: {transcription}")

        # Step 4: Generate speech from the transcribed text using Eleven Labs API
        output_audio_path = os.path.join(download_path, "generated_speech.wav")
        generate_speech_eleven_labs(transcription, output_audio_path)
    else:
        print("Failed to transcribe the audio.")
```
